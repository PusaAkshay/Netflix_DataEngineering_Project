```python
dbutils.widgets.text("sourcefolder","netflix_cast")
dbutils.widgets.text("targetfolder","netflix_cast")
```

```python
var_src_floder = dbutils.widgets.get("sourcefolder")
var_tgt_floder = dbutils.widgets.get("targetfolder")
```

```python
df = spark.read.format("csv")\
    .option("header",True)\
    .option("inferSchema",True)\
    .load(f"abfss://bronze@netflixproject28.dfs.core.windows.net/{var_src_floder}")
```

```python
df = df.write.format("delta")\
.mode("append")\
.option("path",f"abfss://silver@netflixproject28.dfs.core.windows.net/{var_tgt_floder}")\
.save()
```
