```python
%sql
create schema netflix_catalog.netflix_schema
```

```python
checkpoint_location = "abfss://silver@netflixproject28.dfs.core.windows.net/checkpoints_path"
```

```python
df = spark.readStream.format("cloudFiles")\
      .option("cloudFiles.format", "csv")\
      .option("cloudFiles.schemaLocation", checkpoint_location)\
      .load("abfss://raw@netflixproject28.dfs.core.windows.net")
```

```python
df.writeStream.format("delta")\
    .outputMode("append")\
    .option("checkpointLocation", checkpoint_location)\
    .trigger(availableNow=True) \
    .option("path", "abfss://silver@netflixproject28.dfs.core.windows.net/netflix_titles")\
    .start()
```
